```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np. linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
```

```python
fig = plt.figure(figsize=(8,2))
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.6, 0.2, 0.2, 0.2])
axes1.plot(x,y)
axes2.plot(y,x)
                      
```

```python
fig,axe = plt.subplots(figsize = (20,2), nrows = 1, ncols =2)
plt.tight_layout()
axe[0].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color = 'red',lw =1, ls = '--',marker='o', markersize = 15)
axes.set_xlim([0,1])
axes.set_ylim([0,1])
```
