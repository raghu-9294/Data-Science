```python
import seaborn as sns
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde = False, bins = 20)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', palette='coolwarm')
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator = np.std )
```

```python
sns.countplot(x = 'sex', data = tips)
```

```python
sns.boxplot(x='day', y ='total_bill', data = tips)
```

```python
sns.violinplot(x='day',y='total_bill',data = tips,hue ='sex',split = True)
```

```python
sns.stripplot(x='day',y='total_bill',data = tips, hue = 'sex', dodge = True)
```

```python
sns.swarmplot(x='day',y='total_bill',data = tips)
```

```python
sns.catplot(x='day',y='total_bill',data = tips, kind ='strip')
```
